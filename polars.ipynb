```python
df = initialize1() #初期化
df_copy = df.clone()

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder() 
sex_label = le.fit_transform(df_copy["sex"])
emb_label = le.fit_transform(df_copy["embarked"])
df_copy = df_copy.with_columns(
    [
        pl.Series(sex_label).alias("sex"),
        pl.Series(emb_label).alias("embarked")
    ]
)
df_copy.head()
```

```python
df_copy.null_count()
```

```python
df_copy = df_copy.with_columns(
    pl.col(["age","fare"]).fill_null(strategy="mean")
)
df_copy.null_count()
```

```python
df_copy = df_copy.drop(["name","ticket","cabin","boat","body","home.dest"])
df_copy.describe()
```

```python
features = df_copy.select(
    pl.exclude("survived")
).to_numpy()
target = df_copy.select(
    pl.col("survived")
).to_numpy().reshape(-1)
```

```python
from sklearn.model_selection import train_test_split
(features , test_X , target , test_y) = train_test_split(features, target , test_size = 0.3 , random_state = 0)
```

```python
test_X.shape
```

```python
test_y.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100,random_state=0) # ランダムフォレストのインスタンスを作成

model.fit(features,target) # 学習の実行
```

```python
pred = model.predict(test_X)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(pred,test_y)
```

```python
importance = model.feature_importances_ 

print('Feature Importances:')
for i, feat in enumerate(['pclass','age','sex','fare','embarked']):
    print('\t{0:20s} : {1:>.5f}'.format(feat, importance[i]))
```

```python
df_pred = pl.DataFrame(pred)
df_pred.write_csv('submission.csv',has_header=False)
```
